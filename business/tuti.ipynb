```python
import requests
import time
import pandas as pd

# Tu API Key de Google

query = "Tuti"

# URL de la API de Google Places
url = f"https://maps.googleapis.com/maps/api/place/textsearch/json?query={query}&key={API_KEY}"

# Lista para almacenar todos los resultados
all_results = []

# Función para realizar la solicitud y manejar la paginación
def fetch_places(url):
    while True:
        response = requests.get(url)
        data = response.json()
        
        if "results" in data:
            for place in data["results"]:
                name = place.get("name", "Nombre no disponible")
                address = place.get("formatted_address", "Dirección no disponible")
                lat = place["geometry"]["location"]["lat"]
                lng = place["geometry"]["location"]["lng"]

                all_results.append({
                    "Nombre": name,
                    "Dirección": address,
                    "Latitud": lat,
                    "Longitud": lng
                })

                print(f"Nombre: {name}")
                print(f"Dirección: {address}")
                print(f"Latitud: {lat}, Longitud: {lng}")
                print("-" * 40)

        # Verificar si hay más páginas de resultados
        if "next_page_token" in data:
            next_token = data["next_page_token"]
            time.sleep(2)  # Espera requerida para que el token sea válido
            url = f"https://maps.googleapis.com/maps/api/place/textsearch/json?query={query}&key={API_KEY}&pagetoken={next_token}"
        else:
            break

# Ejecutar la búsqueda
fetch_places(url)

# Contar el total de resultados
print(f"Total de tiendas encontradas: {len(all_results)}")

# Guardar los resultados en un archivo CSV
df = pd.DataFrame(all_results)
df.to_csv("tiendas_tuti_ecuador.csv", index=False, encoding="utf-8")

print("Los datos se han guardado en 'tiendas_tuti_ecuador.csv'")
```

```python
import pandas as pd
from rapidfuzz import fuzz, process

# Cargar los archivos CSV


# Elegir las columnas a comparar
column_to_compare_csv1 = "Nombre"
column_to_compare_csv2 = "Nombre"

# Crear listas de nombres
names_csv1 = csv1[column_to_compare_csv1].str.lower().str.strip().tolist()
names_csv2 = csv2[column_to_compare_csv2].str.lower().str.strip().tolist()

# Establecer un umbral de similitud (por ejemplo, 80%)
similarity_threshold = 75

# Lista para almacenar los resultados
matches = []
non_matches = []

# Comparar cada nombre del csv1 con los del csv2
for name in names_csv1:
    match = process.extractOne(name, names_csv2, scorer=fuzz.token_set_ratio)
    if match and match[1] >= similarity_threshold:
        matches.append({
            "Nombre CSV1": name,
            "Nombre CSV2": match[0],
            "Similitud": match[1]
        })
    else:
        non_matches.append({"Nombre CSV1": name})

# Convertir resultados en DataFrames
matches_df = pd.DataFrame(matches)
non_matches_df = pd.DataFrame(non_matches)

# Guardar los resultados en CSV
#matches_df.to_csv("coincidencias_similitud.csv", index=False, encoding="utf-8")
#non_matches_df.to_csv("no_coinciden_similitud.csv", index=False, encoding="utf-8")

# Resumen
print(f"Total coincidencias: {len(matches_df)}")
print(f"Total no coinciden: {len(non_matches_df)}")
```
